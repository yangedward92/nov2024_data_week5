## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_time = df2.groupby("Mouse ID").Timepoint.max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
df3 = pd.merge(max_time, df2, on=["Mouse ID", "Timepoint"], how="left")

drugs = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
mask = df3["Drug Regimen"].isin(drugs)
df3 = df3.loc[mask].reset_index(drop=True)
df3
```

```python
# Change Canvas Size
plt.figure(figsize=(12, 6))

# Create a basic plot
sns.boxplot(data=df3, y="Tumor Volume (mm3)", x="Drug Regimen", color="green")

# Customizations

# Change colors
# Add in reference lines
# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Times New Roman", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Times New Roman", fontsize=12)
plt.title("Drugs Tumor Volume", fontweight="bold", fontsize=16, fontfamily="Times New Roman")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="lightgrey", alpha=0.55)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = df3['Drug Regimen'].unique()

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
# Iterate through each treatment
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment_data = df3.loc[df3['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']

    # add subset
    tumor_volumes.append(treatment_data)
    
    # Calculate IQR for outlier detection
    q1 = np.percentile(treatment_data, 25)
    q3 = np.percentile(treatment_data, 75)
    iqr = q3 - q1
    
    # Determine outliers using upper and lower bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    
    # Identify outliers
    outliers = treatment_data[(treatment_data < lower_bound) | (treatment_data > upper_bound)]
    print(f"Outliers for {treatment}: {outliers.tolist()}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.

# Change Canvas Size
plt.figure(figsize=(12, 6))

plt.boxplot(tumor_volumes, labels=treatments, boxprops=dict(linewidth=2))

# Add Labels/Title
plt.xlabel("Drug Regimen", fontfamily="Times New Roman", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Times New Roman", fontsize=12)
plt.title("Drug Tumor Volume", fontweight="bold", fontsize=16, fontfamily="Times New Roman")

# Add in a background grid
plt.grid(axis="y", linestyle="--", color="lightgrey", alpha=0.75)
# plt.xlim(0,20)

# Show/Save the Graph
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
regimen = "Capomulin"

df5 = df2.loc[df2["Drug Regimen"] == regimen]
df6 = df5.groupby("Mouse ID").agg({"Tumor Volume (mm3)": "mean", "Weight (g)": "mean"}).reset_index()
```

```python
plt.figure(figsize=(12,6))

plt.scatter(df6["Weight (g)"], df6["Tumor Volume (mm3)"], color="green", edgecolor="black", s=200)

plt.xlabel("Weight (g)", fontfamily="Times New Roman", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Times New Roman", fontsize=12)
plt.title("Weight (g) vs Tumor Volume (mm3)", fontweight="bold", fontsize=16, fontfamily="Times New Roman")

plt.grid(linestyle="--", color="lightgrey", alpha=0.55)

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = df6["Weight (g)"]
y_values = df6["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
```

```python
plt.figure(figsize=(12,6))

plt.scatter(df6["Weight (g)"], df6["Tumor Volume (mm3)"], color="green", edgecolor="black", s=200)
plt.plot(df6["Weight (g)"], regress_values, color="black", linewidth=6)


plt.xlabel("Weight (g)", fontfamily="Times New Roman", fontsize=12)
plt.ylabel("Tumor Volume (mm3)", fontfamily="Times New Roman", fontsize=12)
plt.title("Weight (g) vs Tumor Volume (mm3)", fontweight="bold", fontsize=16, fontfamily="Times New Roman")

plt.grid(linestyle="--", color="lightgrey", alpha=0.55)

plt.annotate(line_eq, (23, 36), fontsize=15, color="black")

plt.show()
```
